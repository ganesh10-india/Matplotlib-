# Scatter Plot

# is used to study the behaviour of 2 variables at a time, and also describe the relationship between 2 variables along with direction.

```python
iris.plot.scatter(x='sepal_length', y='sepal_width', title='Iris Dataset')

# Iris is a dataset, i want to plot, and the type of plot is "Scatter Plot"
```

# Line Chart

```python
iris.drop(['species'], axis=1).plot.line(title='Iris Dataset')
```

# Histogram

```python
iris['sepal_length'].plot.hist()
```

```python
iris.plot.hist(subplots=True, layout=(2,2), figsize=(10,10), bins=20)
```

# Bar Chart

```python
iris['sepal_length'].value_counts().sort_index().plot.bar()
```

```python
iris['sepal_length'].value_counts().sort_index().plot.barh()
```

```python
iris.groupby("species").sepal_length.mean().sort_values(ascending=False)[:3].plot,bar()
# group by species(we have 3 levels in species), for sepal compute mean, and sort in descending order
#[:3] means 3 bars, then bar plot
```

# Pie Chart

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data= {'name':['Jason','Mollly','Tina','Jake','Amy'],
           'jan_arrests':[4,24,31,2,3],
           'feb_arrests':[25,94,57,62,70],
           'march_arrests':[5,43,23,23,51]}
df=pd.DataFrame(raw_data,columns = ['name','jan_arrests','feb_arrests','march_arrests'])
df        
```

```python
#Create a column with the total arrests for each officer
df['total_arrests']=df['jan_arrests'] + df['feb_arrests'] + df['march_arrests']
df
```

```python
#create a list of colours (from iWantHue)
colors=['Red','Green','Skyblue','Orange','Yellow','Maroon','White']

#create a pie chart
plt.pie(
    # using data total_arrests
    df['total_arrests'],
    #with the labels being officer names
    labels=df['name'],
    #with colours
    colors=colors,
    #with one slide exploded out
    explode=(0, 0, 0, 0, 0.15),
    #with the start angle at 90%
    startangle=90,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )
    
    
# view the plot drop above
plt.axis('equal')

# view the plot
plt.tight_layout()
plt.show()
```
