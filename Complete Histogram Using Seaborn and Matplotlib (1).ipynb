# Basic histogram | Seaborn

With Seaborn, histograms are made using the distplot function. You can call the function with default values (left), what already gives a nice chart. Do not forget to play with the number of bins using the ‘bins’ argument. It is important to do so: a pattern can be hidden under a bar.

```python
# Import Libraires
import seaborn as sns
import matplotlib.pyplot as plt
df = sns.load_dataset('iris')
# Make default histogram of sepal_lenght
sns.distplot(df['sepal_length'])
plt.show()
```

```python
# Control the number of bins
sns.distplot(df['sepal_length'],bins= 20)
plt.show()
```

# Control rug and density on seaborn histogram
By default, the displot function of seaborn plots an histogram with a density curve. You can easily remove the density using the option kde=”False”. You can also control the presence of rugs using rug=”True”. You can custom rug and density as proposed below:

```python
# Hist only
sns.distplot(a = df['sepal_length'],hist= True,kde=False,rug=False)
```

```python
# Hist + Rug + kernel density
sns.distplot(a = df['sepal_length'],hist=True,kde=True,rug=True)
plt.show()
```

```python
# To change the parameters of rug
sns.distplot(a =df['sepal_length'],rug =True,
            rug_kws={'color':'r','alpha':0.3,'linewidth':2,'height':0.2})
```

```python
# To change parameters of density distribution
sns.distplot( a=df["sepal_length"], kde=True,
kde_kws={"color": "g", "alpha":0.3, "linewidth": 5, "shade":True }
)
```

```python
# Vertical Histogram 
sns.distplot(df['sepal_length'],color='skyblue',vertical= True)
plt.show()
```

# Histogram with a boxplot on top | seaborn

```python
# Cut the window in 2 parts
f, (ax_box,ax_hist) = plt.subplots(2,sharex = True,gridspec_kw={'height_ratios':(.25,.75)})
# add a graph  in each part
sns.boxplot(df['sepal_length'],ax = ax_box)
sns.distplot(df['sepal_length'],ax= ax_hist)

# Remove x -axis name for the boxplot
ax_box.set(xlabel='')
```

# Histogram with several variables | Seaborn

```python
# Method one : on the same axis
sns.distplot(df['sepal_length'],color="skyblue",label='Sepal Length')
sns.distplot(df['sepal_width'],color="green",label='Sepal_Width')
plt.legend()
plt.show()
```

```python
# Method 2 : 
f ,axes =plt.subplots(2,2,figsize=(7,7),sharex=True)
sns.distplot( df["sepal_length"] , color="skyblue", ax=axes[0, 0])
sns.distplot( df["sepal_width"] , color="olive", ax=axes[0, 1])
sns.distplot( df["petal_length"] , color="gold", ax=axes[1, 0])
sns.distplot( df["petal_width"] , color="teal", ax=axes[1, 1])
```

```python
# Marginal plot with Seaborn
# Custom the inside plot: options are: “scatter” | “reg” | “resid” | “kde” | “hex”
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='scatter')
```

```python
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='reg')
```

```python
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='resid')
```

```python
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='kde')
```

```python
sns.jointplot(x=df["sepal_length"], y=df["sepal_width"], kind='hex')
```
